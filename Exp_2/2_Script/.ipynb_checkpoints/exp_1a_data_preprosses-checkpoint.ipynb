```python
def detect_delimiter(csv_file):
    with open(csv_file, 'r') as file:
        dialect = csv.Sniffer().sniff(file.readline())
    return dialect.delimiter
def calcul_stat(data_csv):
    data_list = []
    for k, elem in enumerate(data_csv):
        delimiter = detect_delimiter(elem)
        df = pd.read_csv(elem, header=0, sep=delimiter)
        df_new = df.drop(columns=['fix_start_time', 'dot_mvt_start_time', 'last_dot_time',
                                  'button_pressed_time', 'line_cross_time', 'rest_time', 'end_time',
                                  'start_exp_time', 'time_from_start [ces]', 'early_response (t/f)'])

        # Vectorized operations for new columns
        df_new['prop_last_5'] = (np.sign(df_new['response']) == np.sign(df_new['mean_last5'])).astype(int)
        df_new['prop_last_15'] = (np.sign(df_new['response']) == np.sign(df_new['mean_last15'])).astype(int)
        df_new['prop_ms'] = (np.sign(df_new['response']) == np.sign(df_new['mean_val'])).astype(int)
        df_new['prop_lv'] = (np.sign(df_new['response']) == np.sign(df_new['last_visibl_center_dist[deg]'])).astype(int)
        df_new['RT'] = df_new['RT'].apply(lambda x: x / 1000)

        # Define conditions
        conditions = [
            (df_new['path'] == "5%") & (df_new['last_cond'] == "same"),
            (df_new['path'] == "15%") & (df_new['last_cond'] == "same"),
            (df_new['path'] == "5%") & (df_new['last_cond'] == "opposite"),
            (df_new['path'] == "15%") & (df_new['last_cond'] == "opposite"),
        ]
        choices = ['same_5', 'same_15', 'opposite_5', 'opposite_15']
        condition_values = np.select(conditions, choices, default=None) 
        df_new.insert(4, 'condition', condition_values)
#         df_new['condition'] = np.select(conditions, choices)

        # Calculate accuracy based on conditions
        accuracy_conditions = [
            (df_new['condition'] == 'same_5') | (df_new['condition'] == 'same_15'),
            (df_new['condition'] == 'opposite_5') | (df_new['condition'] == 'opposite_15'),
        ]
        accuracy_choices = [
            (np.sign(df_new['response']) == np.sign(df_new['mean_val'])).astype(int),
            (np.sign(df_new['response']) == np.sign(df_new['mean_last5'])).astype(int),
        ]
        df_new['accuracy'] = np.select(accuracy_conditions, accuracy_choices, default=0)  # Assuming 0 as default for non-specified conditions
        df_new.insert(0, "# subj", np.ones(len(df_new), dtype=int) * k)
        
        data_list.append(df_new)
    return data_list
```

```python
def remove_anomaly_IQR(data_list):
    data_clean_list = []
    for elem in data_list:
        df = elem
        up_same  = df.groupby("last_cond")['RT'].quantile(0.75)[1] +1.5*(df.groupby("last_cond")['RT'].quantile(0.75)[1]-df.groupby("last_cond")['RT'].quantile(0.25)[1])
        up_opposite = df.groupby("last_cond")['RT'].quantile(0.75)[0] +1.5*(df.groupby("last_cond")['RT'].quantile(0.75)[0]-df.groupby("last_cond")['RT'].quantile(0.25)[0])
        
        df.loc[(df['last_cond'] == 'same') & (df['RT'] > up_same), 'RT'] = df.groupby("last_cond")['RT'].median()[1]
        df.loc[(df['last_cond'] == 'opposite') & (df['RT'] > up_opposite), 'RT'] = df.groupby("last_cond")['RT'].median()[0]
        data_clean_list.append(df)
    return data_clean_list   
```

```python
from statsmodels.stats.anova import AnovaRM
mean_rt_df_long = mean_rt_df.stack().reset_index()
mean_rt_df_long.columns = ['subject', 'condition', 'mean_RT']

# Perform ANOVA
# 'subject' represents the subjects, 'condition' represents the different conditions, and 'mean_RT' represents the reaction times
# Change the value of 'subject', 'condition', and 'mean_RT' based on your actual column names
# Here, I'm assuming the subject IDs are in the first column, conditions in the second, and mean RTs in the third column
anovarm_result = AnovaRM(data=mean_rt_df_long, depvar='mean_RT', subject='subject', within=['condition']).fit()

print(anovarm_result.summary())
```
